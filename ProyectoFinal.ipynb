## ***Librerias***

```python
import pandas as pd
import numpy as np

# Visualizacion del arbol
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Regresion Logistica
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

# Evaluacion del Modelo
from sklearn.metrics import accuracy_score, confusion_matrix
```

## ***Limpieza y procesamiento de datos***

> Se carga el excel

```python
df = pd.read_excel("https://github.com/majash29/CienciaDatosSENAV02/raw/main/08_ProyectoFinal/Dataset/10Dataset_medico.xlsx")
df
```

> Se elimina las columnas de Id y Ubicación del dataset

```python
df = df.drop(['Id', 'Ubicacion'], axis=1)
df
```

> Se cambian los nombres de las columnas de género, dolor de garganta, congestión nasal y dificultad respiratoria.

```python
df.rename(columns={'Genero':'Género', 'Dolor de Garganta':'Dolor_de_garganta', 'Congestión Nasal':'Congestión_nasal', 'Dificultad Respiratoria':'Dificultad_respiratoria'}, inplace=True)
df
```

> Se ven los datos de Enfermedad para hacer un map y organizar mejor la clasificación.

```python
print(df['Enfermedad'].unique())
```

> Se realiza la función de map() en algunas columnas para dejar los mismos datos y los diferentes se ponen como nulos por defecto.

```python
df['Fiebre'] = df['Fiebre'].map({1 : 1, 0 : 0})
df['Tos'] = df['Tos'].map({1 : 1, 0 : 0})
df['Dolor_de_garganta'] = df['Dolor_de_garganta'].map({1 : 1, 0 : 0})
df['Congestión_nasal'] = df['Congestión_nasal'].map({1 : 1, 0 : 0})
df['Dificultad_respiratoria'] = df['Dificultad_respiratoria'].map({1 : 1, 0 : 0})
df['Enfermedad'] = df['Enfermedad'].map({'Sinusitis' : 'Sinusitis', 'Gripe' : 'Gripe', 'Resfriado Común' : 'Resfriado_común', 'Resfriado COMUN' : 'Resfriado_común'})

df
```

> Los datos nulos se ponen cero por defecto y los datos de síntomas que estaban como tipo float se pasan a enteros.

```python
df['Fiebre'] = df['Fiebre'].fillna(0)
df['Tos'] = df['Tos'].fillna(0)
df['Dolor_de_garganta'] = df['Dolor_de_garganta'].fillna(0)
df['Congestión_nasal'] = df['Congestión_nasal'].fillna(0)
df['Dificultad_respiratoria'] = df['Dificultad_respiratoria'].fillna(0)
df['Enfermedad'] = df['Enfermedad'].fillna('Resfriado_común')

df['Fiebre'] = df['Fiebre'].astype('int64')
df['Tos'] = df['Tos'].astype('int64')
df['Dolor_de_garganta'] = df['Dolor_de_garganta'].astype('int64')
df['Congestión_nasal'] = df['Congestión_nasal'].astype('int64')
df['Dificultad_respiratoria'] = df['Dificultad_respiratoria'].astype('int64')

df
```

> Se filtra la información de la columna enfermedad para que se cambie a 0 ó 1 los síntomas.

```python
filtro = (df['Enfermedad'] == 'Gripe') | (df['Enfermedad'] == 'Resfriado_común') | (df['Enfermedad'] == 'Sinusitis')
df['Congestión_nasal'] = filtro.replace({False: 0, True: 1})
df['Dificultad_respiratoria'] = filtro.replace({False: 0, True: 1})

filtro2 = (df['Enfermedad'] == 'Gripe') | (df['Enfermedad'] == 'Resfriado_común')
df['Tos'] = filtro2.replace({False: 0, True: 1})

filtro3 = (df['Enfermedad'] == 'Gripe')
df['Fiebre'] = filtro3.replace({False: 0, True: 1})
df['Dolor_de_garganta'] = filtro3.replace({False: 0, True: 1})
```

> Se filtra la información de la columna cliente se cambie a 0 ó 1 el género.

```python
genero = (df['Cliente'] == 'Ana Sánchez') | (df['Cliente'] == 'María Pérez') | (df['Cliente'] == 'Lucía Martínez')
df['Género'] = genero.replace({False: 'Masculino', True: 'Femenino'})
```

> Se guarda un excel con los cambios realizados en la limpieza y el procesamiento de datos.

```python
df.to_excel("Datos_Limbo.xlsx", index=False)
```

## ***Árbol Clasificativo***

### ***División de los datos***

> Se mezclan los datos de dataset

```python
df = shuffle(df, random_state=1)

df
```

> Se seleccionan los datos relavantes para la regresion.

```python
variables = ['Fiebre', 'Tos', 'Dolor_de_garganta', 'Congestión_nasal', 'Dificultad_respiratoria']

X = df[variables]
y = df['Enfermedad']
```

> Dividir el conjunto de datos en entrenamiento y prueba.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

### ***Cronstrucción y aprendizaje del modelo de clasificación***

> Crear el modelo de clasificación.

```python
model = DecisionTreeClassifier(random_state=13)
```

> Entrenar el modelo.

```python
model.fit(X_train, y_train)
```

> Realiza predicción con el conjunto de prueba.

```python
y_pred = model.predict(X_test)
```

### ***Evaluación de modelo de machine learning***

> Calcular la predicción del modelo.

```python
accuracy = accuracy_score(y_test, y_pred)
print('Precisión del modelo: %.2f' % accuracy)
```

> Calcular la matriz de confusión.

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print('Matriz de confusión: ')
conf_matrix
```
